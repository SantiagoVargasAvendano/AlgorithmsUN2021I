### Muestreo
**series.resample(freq)** es una clase llamada "DatetimeIndexResampler" que agrupa los datos de un objeto Series en intervalos de tiempo regulares. El argumento "freq" determina la longitud de cada intervalo.
**series.resample.mean()** es una sentencia completa que agrupa los datos en intervalos y luego calcula la media de cada intervalo. 

```python
por_mes = msft.resample('M').mean()
print(por_mes)
```

También es posible hacerlo por semana, o por cualquier frecuencia que queramos.

```python
por_semana = msft.resample('W').mean()
print(por_semana.head(15))
dos_semanas = msft.resample('2W').mean()
print(dos_semanas.head(15))
```

Tambien podemos usar otro metodos con resampler.

```python
std = msft.resample('M').std()
print(std)
max = msft.resample('M').max()
print(max)
min = msft.resample('M').min()
print(min)
```

A menudo queremos calcular la rentabilidad mensual de una acción, basándonos en los precios del último día de cada mes. Para obtener esos precios, utilizamos el método **series.resample.agg()**

```python
ultimo_dia = msft.resample('M').agg(lambda x: x[-1])
print(ultimo_dia)
```

O bien, calcular directamente las tasas de rendimiento mensuales utilizando los datos del primer día y del último día.

```python
retorno_mensual = msft.resample('M').agg(lambda x: x[-1]/x[1] - 1)
print(retorno_mensual)
```

El objeto Serie también nos proporciona algunos métodos convenientes para hacer algunos cálculos rápidos.

```python
print(retorno_mensual.mean())
print(retorno_mensual.std())
print(retorno_mensual.max())
```

Otros dos métodos utilizados con frecuencia en Series son **.diff()** y **.pct_change()**. El primero calcula la diferencia entre elementos consecutivos, y el segundo calcula el porcentaje de cambio.

```python
print(ultimo_dia.diff())
print(ultimo_dia.pct_change())
```

Cuando se trata de valores NaN, normalmente eliminamos el punto de datos o lo rellenamos con un valor específico. 

```python
retorno_diario = ultimo_dia.pct_change()
print(retorno_diario.fillna(0))
print(retorno_diario.fillna(method = 'bfill'))
```

También podemos eliminar simplemente los valores NaN mediante **.dropna()**.

```python
retorno_diario = ultimo_dia.pct_change().dropna()
print(retorno_diario)
```

### Dataframes
Un DataFrame es una colección de objetos Series, cada uno de los cuales puede contener diferentes tipos de datos. Un DataFrame puede ser creado a partir de varios tipos de datos: diccionario, numpy.ndarray 2-D, una Serie u otro DataFrame.

#### Creando Dataframes
El método más común para crear un DataFrame es pasar un diccionario.

```python
import pandas as pd
import numpy as np
```

```python
dict = {'AAPL': [143.5,  144.09, 142.73, 144.18, 143.77],
        'MSFT': [1023.4,  998.32, 1005.9, 970.12, 982.54],
        'IBM':  [155.58, 153.67, 152.36, 152.94, 153.49]}
dates = pd.date_range('2017-07-03', periods = 5, freq = 'D')
df = pd.DataFrame(dict, index = dates)
print(df)
```

#### Manipulando Dataframes
Podemos obtener valores en un DataFrame por columnas e índices, o si esta no tiene espacios por el nombre

```python
df = msft_table
print(df.Close.tail(10))
print(df['Adj. Volume'].tail(10))
```

Todos los métodos que aplicamos a un índice de Serie como **iloc[ ]**, **loc[ ]** y métodos de remuestreo, también se pueden aplicar a un DataFrame:

```python
msft_2016 = df['2016']
msft_mensual = msft_2016.resample('M').agg(lambda x: x[-1])
print(msft_mensual)
```

Podemos seleccionar determinadas columnas de un DataFrame utilizando sus nombres.

```python
msft_bar = msft_mensual[['Open', 'High', 'Low', 'Close']]
print(msft_bar)
```

Incluso podemos especificar tanto las filas como las columnas utilizando loc[ ]. Los índices de las filas y los nombres de las columnas están separados por una coma

```python
print(msft_mensual.loc['2016-03':'2016-07', ['Open', 'High', 'Low', 'Close']])
```

Los métodos de subconjunto en DataFrame son bastante útiles. Escribiendo declaraciones lógicas entre corchetes, podemos hacer subconjuntos personalizados.

```python
superiores = msft_bar[msft_bar.Close > np.mean(msft_bar.Close)]
print(superiores)
```

#### Validación de datos
Aquí añadimos una nueva columna a un DataFrame existente

```python
msft_bar['rate_return'] = msft_bar.Close.pct_change()
print(msft_bar)
```

Aquí el cálculo introdujo un valor NaN. Si el DataFrame es grande, no podríamos observarlo. **isnull()** proporciona una forma conveniente de comprobar los valores anormales.

```python
faltante = msft_bar.isnull()
print(faltante)
print( '---------------------------------------------')
print(faltante.describe())
```

También podemos encontrar fácilmente las filas con valores perdidos.

```python
print(faltante[faltante.rate_return == True])
```

Como introdujimos en el capítulo de Series, el mismo método **dropna()** y **fillna()** puede aplicarse a un DataFrame.

```python
drop = msft_bar.dropna()
print( drop)
print( '\n--------------------------------------------------\n')
fill = msft_bar.fillna(0)
print( fill)
```

#### Concatenación de Dataframe
En Pandas, la función **concat()** nos permite fusionar múltiples Series en un DataFrame.

```python
s1 = pd.Series([143.5, 144.09, 142.73, 144.18, 143.77], name = 'AAPL')
s2 = pd.Series([898.7, 911.71, 906.69, 918.59, 926.99], name = 'MSFT')
data_frame = pd.concat([s1, s2], axis = 1)
print( data_frame)
```

El parámetro "eje = 1" unirá dos DataFrames por columnas.

```python
log_price = np.log(msft_bar.Close)
log_price.name = 'log_price'
print( log_price)
print( '\n--------------------------------------------\n')
concat = pd.concat([msft_bar, log_price], axis = 1)
print( concat)
```

También podemos unir dos DataFrames por filas.

```python
df_volume = msft_table.loc['2016-10':'2017-04', ['Volume', 'Split Ratio']].resample('2W').agg(lambda x: x[-1])
print( df_volume)
print( '\n-------------------------------------------\n')
df_2017 = msft_table.loc['2016-10':'2017-04', ['Open', 'High', 'Low', 'Close']].resample('2W').agg(lambda x: x[-1])
print( df_2017)
```

Ahora fusionamos los DataFrames con nuestro DataFrame 'msft_bar'

```python
concat = pd.concat([msft_bar, df_volume], axis = 1)
print(concat)
```

Por defecto los DataFrame se unen con todos los datos. Esta opción por defecto hace que la pérdida de información sea nula. También podemos unirlos por intersección, esto se llama 'inner join':

```python
concat = pd.concat([msft_bar, df_volume], axis = 1, join = 'inner')
print( concat)
```

Sólo queda la parte de la intersección si se utiliza el método 'inner join'. Ahora vamos a intentar añadir un DataFrame a otro.

```python
append = msft_bar.append(df_2017)
print( append)
```

'Append' es esencialmente para concatenar dos DataFrames por eje = 0, por lo tanto, aquí es una forma alternativa de anexar.

```python
concat = pd.concat([msft_bar, df_2017], axis = 0)
print( concat)
```

Tenga en cuenta que si los dos DataFrame tienen algunas columnas con los mismos nombres de columna, estas columnas se consideran iguales y se fusionarán. Es muy importante tener los nombres de columna correctos.

```python
df_2017.columns = ['Change', 'High', 'Low', 'Close']
concat = pd.concat([msft_bar, df_2017], axis = 0)
print( concat)
```
