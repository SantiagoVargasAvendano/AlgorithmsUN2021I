### 1. Fibonacci Number

```python
%%writefile FibonacciNumber.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

long int FN(long int n){
    long int f[n+1];
    f[0] = 0;
    f[1] = 1;
    for(long int i=2;i<n+1;i++){
        f[i] = f[i-1] + f[i-2];
    }
    return f[n];
}

int main(){
    long int n;
    cin >> n;
    long int f = FN(n);
    cout << f;
    return 0;
}
```

### 2. Last Digit of a Large Fibonacci Number

```python
%%writefile LastDigitFibonacci.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

long int FN(long int n){
    long int f[61];
    f[0] = 0;
    f[1] = 1;
    for(long int i=2;i<=60;i++){
        f[i] = (f[i-1] + f[i-2])%10;
    }
    return f[n % 60];
}

int main(){
    long int n;
    cin >> n;
    long int f = FN(n);
    cout << f;
    return 0;
}
```

### 3. Greatest Common Divisor

```python
%%writefile GCD.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

int GCD(int a, int b){
    if(b==0){
        return a;
    }else{
        return GCD(b, a%b);
    }
}

int main(){
    int a,b;
    cin >> a;
    cin >> b;
    int gcd = GCD(a,b);
    cout << gcd;
    return 0;
}
```

```python
!g++ GCD.cpp -o GCD
!./GCD
```


### 4. Least Common Multiple

```python
%%writefile LCM.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

long GCD(long a, long b){
    if(b==0){
        return a;
    }else{
        return GCD(b, a%b);
    }
}

int main(){
    long a,b;
    cin >> a;
    cin >> b;
    long lsm = (a*b)/GCD(a,b);
    cout << lsm;
    return 0;
}
```

### 5. Fibonacci Number Again

```python
def fib_mod(n, m):
    
    cache = {0: 0, 1: 1}

    def fib_m(n):
        if n in cache:
            return cache[n]

        if n % 2 == 0:
            fib_half_n = fib_m(n // 2)
            result = (2 * fib_m(n // 2 - 1) + fib_half_n) * fib_half_n
        else:
            result = fib_m((n + 1) // 2) ** 2 + fib_m((n - 1) // 2) ** 2

        cache[n] = result = result % m
        return result

    return fib_m(n)

if __name__ == "__main__":
    n, m = map(int, input().split())
    start_time = time.time()
    print(fib_mod(n, m))
```

### 6. Last Digit of the Sum of Fibonacci Numbers

```python
%%writefile LastDigitSum.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

long int LDSFN(long int n){
    if(n<2){
        return n;
    }else{
        int m = (n+3)%60;
        long int f[m];
        f[0] = 0;
        f[1] = 1;
        for(int i=2;i<m;i++){
            f[i] = (f[i-1] + f[i-2])%10;
        }
        if(f[m-1]==0){
            return 9;
        }else{
            return f[m-1]-1;
        }
    }
}

int main(){
    long int n;
    cin >> n;
    long int f = LDSFN(n);
    cout << f;
    return 0;
}
```

### 8. Last Digit of the Sum of Squares of Fibonacci Numbers

```python
%%writefile LastDigitSquares.cpp
#include <iostream>
#include <stdio.h>
using namespace std;

long int FN(long int n){
    long int f[61];
    f[0] = 0;
    f[1] = 1;
    for(long int i=2;i<=60;i++){
        f[i] = (f[i-1] + f[i-2])%10;
    }
    return f[n % 60];
}

int main(){
    long int n;
    cin >> n;
    long int f = (FN(n)*FN(n+1))%10;  
    cout << f;
    return 0;
}
```
